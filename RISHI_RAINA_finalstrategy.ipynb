```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from datetime import timedelta

plt.style.use("seaborn-v0_8")
np.random.seed(42)
```

```python
STOCKS = [
    "RELIANCE","TCS","INFY","HDFCBANK","ICICIBANK",
    "BHARTIARTL","BAJFINANCE","HINDUNILVR","HCLTECH","MARUTI"
]

def load_stock(symbol):
    df = pd.read_csv(f"{symbol}.csv")
    df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
    df = df.sort_values("Date").set_index("Date")
    df["Stock"] = symbol
    return df

daily = pd.concat([load_stock(s) for s in STOCKS])
```

```python
daily
```

```python
def daily_to_weekly(df):
    weekly = df.resample("W-MON").agg({
        "Open": "first",
        "High": "max",
        "Low": "min",
        "Close": "last",
        "Volume": "sum",
        "Stock": "last"
    })
    weekly["Return"] = weekly["Close"].pct_change()
    return weekly.dropna()

weekly = daily.groupby("Stock", group_keys=False).apply(daily_to_weekly)
```

```python
weekly
```

```python
def add_features(df):
    df["mom_1w"] = df["Close"].pct_change(1)
    df["mom_4w"] = df["Close"].pct_change(4)
    df["mom_12w"] = df["Close"].pct_change(12)

    df["vol_4w"] = df["Return"].rolling(4).std()
    df["vol_12w"] = df["Return"].rolling(12).std()

    df["volume_z"] = (
        (df["Volume"] - df["Volume"].rolling(12).mean())
        / df["Volume"].rolling(12).std()
    )

    return df

weekly = weekly.groupby("Stock", group_keys=False).apply(add_features)
weekly = weekly.dropna()
```

```python
weekly
```

```python
weekly["fwd_return"] = weekly.groupby("Stock")["Return"].shift(-1)
weekly["target"] = (weekly["fwd_return"] > 0).astype(int)
weekly = weekly.dropna()
```

```python
weekly = weekly.sort_index()
train = weekly.loc["2015-01-01":"2022-12-31"]
test  = weekly.loc["2023-01-01":"2025-12-31"]

FEATURES = [
    "mom_1w","mom_4w","mom_12w",
    "vol_4w","vol_12w","volume_z"
]

X_train = train[FEATURES]
y_train = train["target"]

X_test = test[FEATURES]
```

```python
X_train
```

```python
lr = Pipeline([
    ("scaler", StandardScaler()),
    ("lr", LogisticRegression(max_iter=1000))
])

rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=6,
    min_samples_leaf=50,
    random_state=42
)

gb = GradientBoostingClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)

model = VotingClassifier(
    estimators=[("lr", lr), ("rf", rf), ("gb", gb)],
    voting="soft"
)

model.fit(X_train, y_train)
```

```python
test = test.copy()
test["prob_up"] = model.predict_proba(X_test)[:,1]
```

```python
test
```

```python
weekly_selection = (
    test
    .reset_index()
    .sort_values(["Date","prob_up"], ascending=[True,False])
    .groupby("Date")
    .head(2)
)
```

```python
weekly_selection["weight"] = 0.5
weekly_selection[["Date","Stock","prob_up"]].to_csv("predictions.csv", index=False)
weekly_selection[["Date","Stock","weight"]].to_csv("portfolio_weights.csv", index=False)
```

```python
weekly_selection
```

```python
COST = 0.001  # 10 bps per side

weekly_selection["gross_ret"] = weekly_selection["fwd_return"] * weekly_selection["weight"]
portfolio_weekly = weekly_selection.groupby("Date")["gross_ret"].sum()

# Transaction cost: entry + exit
portfolio_weekly_tc = portfolio_weekly -( 2 * COST )
```

```python
def sharpe(returns, freq=52):
    return np.sqrt(freq) * returns.mean() / returns.std()

def max_drawdown(cum):
    peak = cum.cummax()
    return ((cum - peak) / peak).min()

cum = (1 + portfolio_weekly).cumprod()
cum_tc = (1 + portfolio_weekly_tc).cumprod()
```

```python
print("Sharpe (gross):", sharpe(portfolio_weekly))
print("Sharpe (net):", sharpe(portfolio_weekly_tc))
print("Maximum Drawdown (net):", max_drawdown(cum_tc))
```

```python
years = (cum_tc.index[-1] - cum_tc.index[0]).days / 365.25
cagr_net = cum_tc.iloc[-1] ** (1 / years) - 1

print("CAGR (net):", cagr_net)
```

```python
# Annualized Volatility
annual_vol_net = portfolio_weekly_tc.std() * np.sqrt(52)

print("Annualized Volatility (net):", annual_vol_net)
```

```python
# Portfolio Turnover
weights = (
    weekly_selection
    .pivot(index="Date", columns="Stock", values="weight")
    .fillna(0)
)

weekly_turnover = weights.diff().abs().sum(axis=1) / 2
avg_turnover = weekly_turnover.mean()

print("Average Weekly Turnover:", avg_turnover)
```

```python
# Daily returns derived from weekly equity curve
daily_equity = cum_tc.resample("D").ffill()
daily_returns = daily_equity.pct_change().dropna()
```

```python
daily_returns
```

```python
# Monthly Returns
monthly_returns = portfolio_weekly_tc.resample("M").apply(
    lambda x: (1 + x).prod() - 1
)

monthly_returns.to_csv("monthly_returns.csv")
```
